## Reading in the Data

### formating the images

```python
def flat_images(path,label):
#this function reads in images and then flattens them into a dataframe
    images=imread_collection(path)
    flat_images=[]
    for i in range(len(images)):
        img=images[i]
        img1=img.flatten()
        flat_images.append(img1)
    length=flat_images[1].shape
    df=pd.DataFrame(flat_images)
    df['label']=label
    return(df)
```

```python
#specifies path and label for pictures of Austin
path = 'C:/Users/sam2146/Documents/522_Proj/small/*.jpg'
label=[1,2,3]
df=flat_images(path,label[0])
```

```python
#Adds a position label so we can tell what features are in the pictures
position=[np.floor(i/7) for i in df.index]
df['position']=position
```

### PCA reduction on the Data Frame

```python
#Drop the label and position column to run the pca
nonlabel=big_df.drop(['label','position'],axis=1)
labels=big_df[['label','position']]
```

```python
#PCA on the unlabeled dataframe
pca=PCA()
pca.fit(nonlabel)
csum=np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(csum>=.9999)+1
d
```

```python
#New dataframe shape
pca=PCA(n_components=d)
big_df_reduced=pca.fit_transform(nonlabel)
big_df.shape, big_df_reduced.shape
```

```python
#adding the labels back to the reduced dataframe
pca_flat=pd.DataFrame(big_df_reduced)
pca_flat['label']=labels.label.values
pca_flat['position'] = labels.position.values
pca_flat.head()
```

## MLP Classifier

### Scaling

```python
#Scaling the data
X=pca_flat.drop(['position','label'],axis=1)
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
#creating a train test split
y=pca_flat['label']
strat=pca_flat[['position','label']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.285,stratify=strat,random_state=100)
```

```python
#X shapes
print(X_train.shape)
print(X_test.shape)
```

### Gridsearch

```python
#Gridsearch for the MLP Classifer
params={'max_iter':[500],'alpha':[.00001,.001,1],'hidden_layer_sizes':[(100,100,100),(10,10,10)],'random_state':[42]}
grid=GridSearchCV(MLPClassifier(),params,refit=True,verbose=3,cv=5)
grid.fit(X_train,y_train)
```

```python
#Use the test data to get predictions
y_pred=grid.predict(X_test)
```

### Results

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
